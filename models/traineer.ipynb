# Imports

```python
from features.antispoof import get_full_feature_vector
import cv2 
import torch
import  torch.optim as optim 
from torch.nn import nn 
```

# Loading Data

```python
path = r"D:\Storeage-1\Main\ML-Model\data_set\pandu.mp4"



def genData(path):
    feature = []
# Image / Video Loading 
    image_extensions = {".jpg", ".png", ".jpeg", ".webp"}
    if (path.split(".")[1]in image_extensions):
        features = get_full_feature_vector(path,flg=True)
    else:
        cap = cv2.VideoCapture(path)
        while cap.isOpened():
            ret,frame = cap.read()
            if not ret:
                break
            offs = get_full_feature_vector(frame,flg=False)
            feature.append(offs)

    
    return feature


```

# Initialize the Neural Network

```python

# Neuarl Network 

class Markov(nn.Module):
    def __init__(self,ips):
        super().__init__()

        self.fc1 = nn.Linear(ips,1024)
        self.fc2 = nn.Linear(1024,512)
        self.fc3 = nn.Linear(512,256)
        self.fc4 = nn.Linear(256,8)
        self.fc5= nn.Linear(8,2)
        self.sm = nn.Softmax(dim=1)
        self.relu = nn.ReLU()
        


    def forward(self ,x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.relu(self.fc3(x))
        x = self.relu(self.fc4(x))
        x = self.sm(self.fc5(x))

        return x 
    

size = len(genData(path))
model = Markov(size)


criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```
